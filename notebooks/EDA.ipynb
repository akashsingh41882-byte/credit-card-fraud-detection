# Exploratory Data Analysis — Credit Card Fraud Detection

Quick look at class balance, amount distributions, and feature separability.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/creditcard.csv')
df.head()
```

```python
# Class balance
counts = df['Class'].value_counts()
print(counts)
print(f"Fraud rate: {df['Class'].mean() * 100:.3f}%")

counts.plot(kind='bar', title='Class distribution (0 = Legit, 1 = Fraud)')
plt.show()
```

```python
# Transaction amount by class
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(df[df['Class'] == 0]['Amount'], bins=50, ax=axes[0])
axes[0].set_title('Amount - Legit')
sns.histplot(df[df['Class'] == 1]['Amount'], bins=50, ax=axes[1], color='red')
axes[1].set_title('Amount - Fraud')
plt.tight_layout()
plt.show()
```

```python
# Correlation of features with the target
corr_with_target = df.corr(numeric_only=True)['Class'].drop('Class').sort_values()
corr_with_target.plot(kind='barh', figsize=(6, 8), title='Feature correlation with Class')
plt.show()
```
